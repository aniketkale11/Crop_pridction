```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cp=pd.read_csv('cp.csv')
cp.head()
```

```python
cp.info()
```

```python
cp.isnull().sum()
```

```python
cp.columns
```

```python
X=cp[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]
Y=cp["label"]
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,7))
sns.heatmap(X.corr(),annot=True,cmap="viridis")
ax.set (ylabel="features")
ax.set(xlabel="features")
plt.title('correlation between differnt feature')
plt.show()
```

```python
cp['label'].value_counts()
```

```python
pd.pivot_table(cp,index='label',values=['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
acc=[]
model=[]
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
import warnings
warnings.filterwarnings("ignore")
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print (classification_report(Y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(model,X,Y,cv=5)
print ("cross val score is ",score)
```

```python
from flask import Flask
from flask import render_template
import joblib
from flask import request
```

```python
app=Flask(__name__)
model=joblib.load('crop.pkl')
@app.route('/')
def hello(name=None):
    return render_template('index2.html',name=name)

@app.route('/result',methods=['GET','POST'])
def get_value():
    if request.method=="POST":
        h=float(request.form.get('h'))
        p=int(request.form.get('p'))
        K=int(request.form.get('K'))
        temp=int(request.form.get('temp'))
        hue=int(request.form.get('hue'))
        ph=int(request.form.get('ph'))
        rain=int(request.form.get('rain'))
        predict=model.predict([[h,p,K,temp,hue,ph,rain]])
        print(predict[0])
    
    return render_template('result.html',prediction=[predict[0]])
    
```

```python
if __name__=="__main__":
    app.run()
```
